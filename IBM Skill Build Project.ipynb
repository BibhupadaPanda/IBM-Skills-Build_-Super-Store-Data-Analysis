```python
# Visualize sales by category
plt.figure(figsize=(10, 6))
sns.barplot(x="Category", y="Sales", data=df)
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Visualize sales by sub-category
plt.figure(figsize=(12, 6))
sns.barplot(x="Sub-Category", y="Sales", data=df)
plt.title("Total Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize profit by category
plt.figure(figsize=(10, 6))
sns.barplot(x="Category", y="Profit", data=df)
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
# Visualize profit by sub-category
plt.figure(figsize=(12, 6))
sns.barplot(x="Sub-Category", y="Profit", data=df)
plt.title("Total Profit by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
# Group the data by 'Sub-Category' and calculate total profit
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index()

# Filter out sub-categories with negative profit
profit_by_subcategory = profit_by_subcategory[profit_by_subcategory['Profit'] >= 0]

# Sort the data in descending order of total profit
profit_by_subcategory = profit_by_subcategory.sort_values(by='Profit', ascending=False)

# Select the top 10 profitable sub-categories
top_10_profitable_subcategories = profit_by_subcategory.head(10)

# Create a pie chart for top 10 total profit by sub-category using matplotlib
plt.figure(figsize=(10, 8))
plt.pie(top_10_profitable_subcategories['Profit'], labels=top_10_profitable_subcategories['Sub-Category'], autopct='%1.1f%%', startangle=140)
plt.title("Top 10 Total Profit by Sub-Category")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Visualize profit by region
plt.figure(figsize=(8, 6))
sns.barplot(x="Region", y="Profit", data=df)
plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.show()
```

```python
# Visualize top-selling products
top_products = df.groupby("Sub-Category")["Quantity"].sum().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_products.index, y=top_products.values)
plt.title("Top 10 Selling Products")
plt.xlabel("Product Name")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

```python
# Group the data by 'Sub-Category' and calculate total profit
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index()

# Sort the data in ascending order of total profit to find sub-categories with high loss
loss_by_subcategory = profit_by_subcategory[profit_by_subcategory['Profit'] < 0]

# Create a bar plot for sub-categories with high loss
plt.figure(figsize=(10, 6))
plt.bar(loss_by_subcategory['Sub-Category'], loss_by_subcategory['Profit'], color='skyblue')
plt.title("Sub-Categories with High Loss")
plt.xlabel("Sub-Category")
plt.ylabel("Total Loss")
plt.xticks(rotation=90)
plt.show()
```
